```python
f_img = '/Users/kuba/Documents/data/Raw/Face-Detection-Dataset/images/train/0a0d7a87378422e3.jpg'
f_label = '/Users/kuba/Documents/data/Raw/Face-Detection-Dataset/labels/train/0a0d7a87378422e3.txt'
```

```python
with open(f_img, 'rb') as pgmf:
    im = plt.imread(pgmf)
    plt.imshow(im)
```

```python
#now lets get the box on her face
lables = np.loadtxt(f_label)

print(f'Class: {lables[0]} \nNormalized Center X coordinate: {lables[1]} \nNormalized Center Y coordinate: {lables[2]} \nNormalized width: {lables[3]} \nNormalized height: {lables[4]}')
```

```python
with open(f_img, 'rb') as pgmf:
    im = plt.imread(pgmf)

center_x_pxl = int(len(im[0]) * lables[1])
center_y_pxl = int(len(im) * lables[2])
print(f'The center pxl value is at ({center_x_pxl},{center_y_pxl})')
```

```python
plt.imshow(im)
plt.plot(center_x_pxl, center_y_pxl, 'bo')
plt.show()
```

```python
#now lets get the bounding box on the img

#horizonal lines
y0 = center_y_pxl + int(len(im) *  (lables[4] / 2))
y1 = center_y_pxl - int(len(im) *  (lables[4] / 2))
xmin = center_x_pxl - int(len(im[0]) *  (lables[3] / 2))
xmax = center_x_pxl + int(len(im[0]) *  (lables[3] / 2))
print(f" y0:{y0} y1:{y1} xmin:{xmin} xmax:{xmax}")

#vertical lines
x0 = center_x_pxl + int(len(im[0]) *  (lables[3] / 2))
x1 = center_x_pxl - int(len(im[0]) *  (lables[3] / 2))
ymin = center_y_pxl - int(len(im) *  (lables[4] / 2))
ymax = center_y_pxl + int(len(im) *  (lables[4] / 2))

print(f" x0:{x0} x1:{x1} ymin:{ymin} ymax:{ymax}")
```

```python
plt.imshow(im)
plt.plot(center_x_pxl, center_y_pxl, 'ro')
plt.hlines(y0,xmin=xmin,xmax=xmax, colors='red')
plt.hlines(y1,xmin=xmin,xmax=xmax, colors='red')

plt.vlines(x0, ymin, ymax, colors='red')
plt.vlines(x1, ymin, ymax, colors='red')

plt.show()
```
