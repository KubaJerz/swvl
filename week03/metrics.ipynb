### For loading metrics from json and reploting

```python
import json
import matplotlib.pyplot as plt
```

```python
with open('/home/kuba/projects/swvl/week03/experiments/try15_adagrad/metrics.json','r') as file:
    dic = json.load(file)
```

```python
lossi, devlossi = dic['lossi'], dic['devlossi']
```

```python
plt.plot(lossi[1:], label='lossi')
plt.plot(devlossi[1:], label='devlossi')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.ylim((0,0.1))
# plt.savefig(SAVE_DIR+'/'+NAME+"/"+"loss.png")
# plt.close()
```
